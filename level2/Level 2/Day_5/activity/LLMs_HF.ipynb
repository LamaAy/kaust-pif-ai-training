```python
#@title Simple inference of an LLM from HF
# HF library transformers docs -> https://huggingface.co/docs/transformers/en/index
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# ---- Simple Config (students only change these) ----
# MODEL_ID = "Qwen/Qwen3-1.7B"
MODEL_ID = "Qwen/Qwen2.5-1.5B-Instruct"

VARIABLE_DETAIL = "Explain step by step like a teacher."

# ---- Build prompt ----
prompt = f"""
You are a helpful reasoning assistant.
{VARIABLE_DETAIL}
And only answer the question precisely and concisely.

The Question:
Why do airplanes stay in the air?

Answer:
"""

# ---- Load model ----
device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.bfloat16 if torch.cuda.is_available() else torch.float32

tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    torch_dtype=dtype,
    device_map="auto"
)

# ---- Run inference (mostly defaults) ----
inputs = tokenizer(prompt, return_tensors="pt").to(device)
output = model.generate(**inputs, max_new_tokens=256)

print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
#@title choose a model to prompt and try: (qwen2, qwen3, gemma or phi)
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# =========================================
# choose ONE option here
# "qwen2" | "qwen3" | "gemma" | "phi"
# =========================================
MODEL_CHOICE = "qwen2"  #@param ["qwen2", "qwen3", "gemma", "phi"]

MODEL_MAP = {
    "qwen2": "Qwen/Qwen2.5-1.5B-Instruct",
    "qwen3": "Qwen/Qwen3-1.7B",
    "gemma": "google/gemma-3-7b-it",
    "phi":   "microsoft/Phi-3-mini-4k-instruct"
}

MODEL_ID = MODEL_MAP[MODEL_CHOICE]

print(f"Loading: {MODEL_ID}")

# =========================================
# ðŸš€ Load Model + Tokenizer
# =========================================
device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.bfloat16 if torch.cuda.is_available() else torch.float32

tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    torch_dtype=dtype,
    device_map="auto"
)

# =========================================
# ðŸ§  Simple Inference Function
# Students will call ONLY this
# =========================================
def run_model(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    output = model.generate(
        **inputs,
        max_new_tokens=256
    )
    text = tokenizer.decode(output[0], skip_special_tokens=True)
    print("\n===== MODEL OUTPUT =====\n")
    print(text)
```

## **Retrieval Augmented Generation (RAGs)**

```python
#@title Load & Chunk the PDF
from pypdf import PdfReader

PDF_PATH = "/content/CV-example.pdf"

def load_pdf(path):
    reader = PdfReader(path)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text

raw_text = load_pdf(PDF_PATH)
print("PDF Loaded. Characters:", len(raw_text))
```

```python
#@title Create Chunks + Build Retriever
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss

CHUNK_SIZE = 600
OVERLAP = 100

def chunk_text(text):
    chunks = []
    start = 0
    while start < len(text):
        end = start + CHUNK_SIZE
        chunks.append(text[start:end])
        start += CHUNK_SIZE - OVERLAP
    return chunks

chunks = chunk_text(raw_text)
print("Chunks:", len(chunks))

# embed
embedder = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
embeddings = embedder.encode(chunks)

# build FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))
```

```python
#@title Retrieve Relevant Chunks
def retrieve(query, k=3):
    q_emb = embedder.encode([query])
    distances, idx = index.search(np.array(q_emb), k)
    return [chunks[i] for i in idx[0]]
```

```python
MODEL_ID = "Qwen/Qwen3-1.7B"

device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.bfloat16 if torch.cuda.is_available() else torch.float32

tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    torch_dtype=dtype,
    device_map="auto"
)
```

```python
#@title RAG Query function CV specialist LLM
# If you want to make this LLM to extract something else in the CV change the prompt.
def rag_query(question):
    retrieved = retrieve(question, k=4)
    context = "\n\n".join(retrieved)

    prompt = f"""
You are an expert resume analyzer.
You are given content from a candidate CV.

Context:
{context}

Task:
Extract the KEY SKILLS of this candidate.
Summarize them CLEARLY in 5 short bullet lines.
Avoid unrelated info.

#######################################################

LLM's Answer:
"""

    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    output = model.generate(
        **inputs,
        max_new_tokens=300,
        temperature=0.3,
        top_p=0.9
    )

    print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
rag_query("What key skills does this candidate have?")
```
